<h1>Majurca: Model performance visualization</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_model(model):
    df = pd.read_csv(
        model_dir + "model-" + model + "/history.csv"
    )
    df.name = model
    return df

def display_model(df):
#     yticks = np.arange(df["val_accuracy"].min(), df["val_accuracy"].max(), 
#                        0.001)
    yticks = np.arange(0, 1.05, 0.05)
    
    xticks = np.arange(0, len(df), 10)
    
    df[["accuracy", "val_accuracy"]].plot(grid=True, figsize=(15,8), xticks=xticks, 
                                          yticks=yticks)
    plt.show()
    
    print("accuracy", round(df.iloc[-1]["accuracy"],4))
    print("val_accuracy", round(df.iloc[-1]["val_accuracy"],4))
    
def compare_metric(models, metric):
    plt.figure(figsize=(15,8))
    
    for df in models:
        plt.plot(df[metric], label=df.name)
        plt.legend()
    plt.show()
```

```python
model_dir = "/home/otiose/repos/epita/majurca/model/"
```

<h2>Model 01</h2>

```python
df01 = load_model("01")
display_model(df01)
```

<h2>Model 02</h2>

```python
df02 = load_model("02")
display_model(df02)
```

```python
model_df = [df01, df02]
```

<h2>Model 01 & 02 training accuracy comparision</h2>

```python
compare_metric(model_df, "accuracy")
```

<h2>Model 01 & 02 validation accuracy comparison</h2>

```python
compare_metric(model_df, "val_accuracy")
```
